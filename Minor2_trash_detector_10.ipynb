```python
# Step 1: Install YOLOv8 (Ultralytics)
!pip install ultralytics --upgrade --quiet

# Step 2: Import required libraries
import os
from google.colab import files
import zipfile

# Step 3: Upload your dataset ZIP (the same you shared earlier)
uploaded = files.upload()

# Step 4: Unzip the dataset
for fname in uploaded.keys():
    if fname.endswith('.zip'):
        with zipfile.ZipFile(fname, 'r') as zip_ref:
            zip_ref.extractall('/content/dataset')

# Step 5: Check the extracted structure
!ls /content/dataset

# Step 6: Train the YOLOv8 model
from ultralytics import YOLO

# Create the model (use yolov8n, yolov8s, yolov8m, yolov8l depending on your GPU)
model = YOLO('yolov8n.pt')  # 'n' is nano, fastest and lightest

# Train the model
results = model.train(
    data='/content/dataset/data.yaml',  # Path to your YAML
    epochs=50,                          # You can increase if needed
    imgsz=640,                          # Image size
    batch=16,                           # Adjust based on GPU memory
    name='yolov8n-trash',              # Experiment name
)
```
